# Part 1: Basic grouping

```python
#q25
q25_df=pd.read_sql("""
SELECT COUNT(TrackId) AS TrackCount, GenreId
FROM Track
GROUP BY GenreId 
ORDER BY TrackCount DESC, GenreId ASC
""",conn)
q25_df.to_pickle("q25.pkl")
```

```python
q25_df
```

```python
#q26
q26_df=pd.read_sql("""
SELECT SUM(Milliseconds) / 3600000 AS TotalDurationHours, Track.AlbumId, Album.Title AS AlbumTitle
FROM Track
INNER JOIN Album ON Track.AlbumId = Album.AlbumId
GROUP BY Album.AlbumId, Album.Title
ORDER BY TotalDurationHours DESC
LIMIT 5
""",conn)
q26_df.to_pickle("q26.pkl")
```

```python
q26_df
```

```python
#q27
q27_df=pd.read_sql("""
SELECT COUNT(DISTINCT GenreId) AS GenreCount, Album.Title AS AlbumTitle
FROM Track
INNER JOIN Album ON Track.AlbumId = Album.AlbumId
GROUP BY Album.AlbumId,AlbumTitle
HAVING GenreCount > 1
ORDER BY GenreCount DESC, AlbumTitle ASC
""",conn)
q27_df.to_pickle("q27.pkl")
```

```python
q27_df
```

```python
#q28
q28_df=pd.read_sql("""
SELECT Artist.Name as ArtistName,SUM(InvoiceLine.UnitPrice*InvoiceLine.Quantity) AS TotalRevenue
FROM InvoiceLine
INNER JOIN Track ON InvoiceLine.TrackId = Track.TrackId
INNER JOIN Album ON Track.AlbumId = Album.AlbumId
INNER JOIN Artist ON Album.ArtistId=Artist.ArtistId
GROUP BY Artist.ArtistId,ArtistName
ORDER BY TotalRevenue DESC, ArtistName ASC
""",conn)
q28_df.to_pickle("q28.pkl")
```

```python
q28_df
```

```python
#q29
q29_df=pd.read_sql("""
SELECT Genre.Name as GenreName,AVG(Track.Milliseconds/60000) AS AverageDurationMinutes
FROM Genre
INNER JOIN Track ON Genre.GenreId = Track.GenreId
GROUP BY Genre.GenreId, GenreName
HAVING AverageDurationMinutes>20
ORDER BY AverageDurationMinutes DESC
""",conn)
q29_df.to_pickle("q29.pkl")
```

```python
q29_df
```

```python
#q30
q30_df=pd.read_sql("""
SELECT Customer.CustomerId, FirstName, LastName, SUM(InvoiceLine.UnitPrice*InvoiceLine.Quantity) AS TotalExpenditure, RANK() OVER (ORDER BY SUM(InvoiceLine.UnitPrice * InvoiceLine.Quantity) DESC) AS ExpenditureRank
FROM Customer
INNER JOIN Invoice ON Customer.CustomerId=Invoice.CustomerId
INNER JOIN InvoiceLine ON Invoice.InvoiceId=InvoiceLine.InvoiceId
INNER JOIN Track ON InvoiceLine.TrackId = Track.TrackId
INNER JOIN Album ON Track.AlbumId = Album.AlbumId
INNER JOIN Artist ON Album.ArtistId = Artist.ArtistId
WHERE Artist.Name = 'Queen'
GROUP BY Customer.CustomerId, FirstName, LastName
ORDER BY TotalExpenditure DESC, Customer.CustomerId ASC
""",conn)
q30_df.to_pickle("q30.pkl")
```

# Part 2: Windowing queries

```python
#q31
q31_df=pd.read_sql("""
SELECT 
    TrackId, 
    Name AS TrackName, 
    Milliseconds / 60000.0 AS DurationMinutes, 
    RANK() OVER (ORDER BY Milliseconds DESC, Milliseconds / 60000.0 DESC) AS DurationRank
FROM Track
ORDER BY DurationRank ASC, TrackId ASC
""",conn)
q31_df.to_pickle("q31.pkl")
```

```python
q31_df
```

```python
#q32
q32_df=pd.read_sql("""
SELECT Customer.CustomerId, FirstName, LastName, SUM(InvoiceLine.UnitPrice*InvoiceLine.Quantity) AS TotalExpenditure, RANK() OVER (ORDER BY SUM(InvoiceLine.UnitPrice * InvoiceLine.Quantity) DESC) AS ExpenditureRank
FROM Customer
INNER JOIN Invoice ON Customer.CustomerId=Invoice.CustomerId
INNER JOIN InvoiceLine ON Invoice.InvoiceId=InvoiceLine.InvoiceId
INNER JOIN Track ON InvoiceLine.TrackId = Track.TrackId
INNER JOIN Album ON Track.AlbumId = Album.AlbumId
INNER JOIN Artist ON Album.ArtistId = Artist.ArtistId
WHERE Artist.Name = 'Queen'
GROUP BY Customer.CustomerId, FirstName, LastName
ORDER BY TotalExpenditure DESC, Customer.CustomerId ASC
""",conn)
q32_df.to_pickle("q32.pkl")
```

```python
q32_df
```

```python
#q33
q33_df=pd.read_sql("""

SELECT Customer.CustomerId, FirstName, LastName, COUNT(DISTINCT Invoice.InvoiceId) AS InvoicesCount, 
       ROW_NUMBER() OVER (ORDER BY COUNT(DISTINCT Invoice.InvoiceId) DESC, LastName ASC) AS InvoiceRank
FROM Customer
INNER JOIN Invoice ON Customer.CustomerId=Invoice.CustomerId
GROUP BY Customer.CustomerId, FirstName, LastName
""",conn)
q33_df.to_pickle("q33.pkl")
```

```python
q33_df
```

```python
#q34
q34_df= pd.read_sql("""
SELECT * FROM (
SELECT BillingCountry, InvoiceId, Total, RANK() OVER (PARTITION BY BillingCountry ORDER BY Total DESC) AS InvoiceRank
FROM Invoice)TopThree
WHERE InvoiceRank<=3
ORDER BY BillingCountry ASC, InvoiceRank ASC, InvoiceId ASC
""",conn)
q34_df.to_pickle("q34.pkl")
```

```python
q34_df
```

```python
#q35
q35_df= pd.read_sql("""
SELECT Month, MonthlySales, AVG(MonthlySales) OVER(ROWS BETWEEN 2 PRECEDING AND CURRENT ROW) AS MovingAverageSales
FROM(
SELECT DATE_FORMAT(InvoiceDate, '%Y-%m') AS Month,  SUM(Total) AS MonthlySales
FROM Invoice
GROUP BY Month
)TotalSales
ORDER BY Month ASC
""",conn)
q35_df.to_pickle("q35.pkl")
```
